#  Hangman

Two player game. Player 1 enters the word and Player 2 needs to guess the word.

Game explains if you have used the letter already and whether the letter is correct or wrong.

If wrong the game will deduct one guess. The game ends when Player 2 has guessed the correct word or has no guesses left.

Example given below.

```python
def position(lst, item):
    return [i for i, x in enumerate(lst) if x == item]
guesses_left=10
import time
import getpass
#Start of Game
word = getpass.getpass("Player 1: Enter a word")
word_list=list(word)
lol=[]
guessed_letters=[]
wrong_letters=[]

for i in word_list:
    lol=lol+["?"]
    hangman="".join(lol)
print("Starting Game...")
time.sleep(3)
print(hangman, "| ","Guesses left: ",guesses_left)
print("Player 2 - Guess a letter")

#loop till game ends
while word!=hangman:
    guess=input()
    
    if guess in guessed_letters:
        print("You already tried this letter!")
        print(hangman, "|", "Guesses left: ",guesses_left)
    else:
        if guess in word_list:
            a=position(word_list, guess)
            print("Correct!") 
            print("")
            for number in a:
                hangman=hangman[:number]+guess+hangman[number+1:]
            print(hangman, "|", "Guesses left: ",guesses_left)
            if word==hangman:
                break  
            guessed_letters=guessed_letters+[guess]

        else:        
            guesses_left=guesses_left-1
            if guesses_left==0:
                break 
            print("Wrong, try again!")
            print("")
            print(hangman, "|", "Guesses left: ",guesses_left)
            guessed_letters=guessed_letters+[guess]
            wrong_letters=wrong_letters+[guess]
            
        print("Guessed Letters:",guessed_letters)
        print("Wrong Letters:",wrong_letters)
    
    
    

    
if guesses_left==0:
    print("Sorry you lost! The word was: ", word)
if guesses_left>0:
    print("Congratulations, you won!")
```
