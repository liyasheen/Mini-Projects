Dice Game
This game involves 2 players creating a username and password.
The 2 players each roll a pair of dice for a total of 3 rounds.
The points achieved depends on if the total of the dice is odd, even or doubles.
The total number of points is calculated for each player and the winner is announced at the end.

```python
username=input('Player 1 Create username: ')
password=input('Player 1 Create password: ')
```

```python
username2=input('Player 2 Create username: ')
password2=input('Player 2 Create password: ')
```

```python
u_login=input('Player 1 Enter username: ')
if u_login==username:
    p_login=input('Player 1 Enter password: ')
    if p_login==password:
        print('Login Sucess!')
    else:
        print('Wrong Password')
else:
    print('No User found')
```
