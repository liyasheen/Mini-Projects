A standalone program that when executed prompts the user for the name of an XYZ file containing a structure with an iron atom, and after the user enters this name, the program prints out a list with the atoms bonded to the iron atom.

```python
#Turns List into a float
def tofloat (lst):
    new_list = [float(i) for i in lst]
    return new_list
```

```python
#Function receives one string as its argument (a line of the XYZ file) and returns a list of the form [element_symbol,coordinate_list]
def parse_line(each_line):
        list1=[]
        for line in each_line:
            coordinates=tofloat(each_line.split()[1:])
            list1=[each_line.split()[0]] + [coordinates]
        return list1
```

```python
#Function returns item in list
def formalist(lst): 
    return [[el] for el in lst]
```

```python
#Function formats the file into list form
def form_list (name_of_file):
    file_list=[]
    new_list=[]
    with open(name_of_file) as f:
        mylist = list(f)
        for item in mylist:
            file_list=file_list+[item]
    list00=formalist(file_list)
    for k in list00[2:]:
        for l in k:
            new_list=new_list+[parse_line(l)]
    return new_list


```

```python
#This function receives a list, where each element of the list is of the same form as the output of the function parse_line(). This function returns the element of the input list which has the form:["Fe",coordinate_list]
def find_iron(enter_list):
    for item in enter_list:
        if 'Fe' in item:
            return item
```

```python
#This function receives two lists as arguments, each list containing the coordinates of a point in 3D space. The function returns the distance between these two points in space.
def distance(list1, list2):
    a=(list1[0]-list2[0])**2
    b=(list1[1]-list2[1])**2
    c=(list1[2]-list2[2])**2
    distance_between_points=(a+b+c)**0.5
    return distance_between_points
```

```python
from cov_radii import *
cov_list
def return_cov_radius(element):
    for item in cov_list:
        if element in item:
            radius=item[1]
    return radius
```

```python
#This function receives two lists as arguments, each in the same form as the output of function parse_line(). Using the information about the covalent radius of the atoms, it returns the boolean True if the two atoms represented by list1 and list2 are at a distance which is smaller than the sum of their covalent radii with a tolerance of 0.4 Ã…. It returns False if that is not the case.
def is_bonded(list1,list2):
    atom1=list1[1]
    atom2=list2[1]
    interatomic=distance(atom1,atom2)
    cov_radius=return_cov_radius(list1[0]) + return_cov_radius(list2[0])
    if interatomic-cov_radius>0.4:
        return False
    if interatomic-cov_radius<0.4:
        return True
```

```python
#Returns a list of atoms bonded to iron
def atom_bonded(atom_list):
    list_of_atoms=[]
    for item in atom_list:
        true_or_false=is_bonded(item,Fe_list)
        if true_or_false is True:
            list_of_atoms=list_of_atoms+[item]
    return list_of_atoms
```
